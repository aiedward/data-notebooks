```python
# import packages

import pandas as pd
import numpy as np
import pickle
from collections import Counter
from itertools import combinations
from pprint import pprint


# read pre-trained NER data

df = pd.read_csv('./NER_study_abroad.csv')

with open('./NER_list_1-20000.pickle', 'rb') as file:
    enr_list1 = pickle.load( file )
    file.close()

with open('./NER_list_20001-40000.pickle', 'rb') as file:
    enr_list2 = pickle.load( file )
    file.close()

with open('./NER_list_40000-.pickle', 'rb') as file:
    enr_list3 = pickle.load( file )
    file.close()

    
enr_list = enr_list1 + enr_list2 + enr_list3

co_ocurr = []


# get entity that relevent
# as for the entity code meaning, see: https://wiki.planetoid.info/index.php/Named_entity_recognition_tools

for e in enr_list:
    
    lst = []
    
    for i in e:
        # add code if needed
        if i[2] in ['ORG', 'GPE', 'NORP', 'FAC', 'LOC']:
            lst.append(i[3])
            
    if lst != []:
        co_ocurr.append(lst)
        
        
tmp = []
for i in co_ocurr:
    tmp += i
    
types = list(set(tmp))

N = len(types)
matrix = { i:{} for i in types }


# build co-occurence dictionary

for i in co_ocurr:
#     print(i)
    if len(i) > 1:
        comb = [j for j in combinations(i, 2)]
        for item in comb:
            
            # count co-occurrence
            if item[1] in matrix[item[0]]: # key, dict
                matrix[item[0]][item[1]] += 1
            else:
                matrix[item[0]][item[1]] = 1
                
            if item[0] in matrix[item[1]]: # key, dict
                matrix[item[1]][item[0]] += 1
            else:
                matrix[item[1]][item[0]] = 1
                
sort_keys = sorted(matrix, key=lambda x: len(matrix[x]), reverse=True)

# print entity names

print('Entity List:')
pprint(sort_keys)
```

```python
while True:
    query = input('type "stop" to break, or enter the entity name: ')

    if query == 'stop':
        print('thanks for watching!')
        break

    tmp = [(i, matrix[query][i]) for i in sorted( matrix[query], key=lambda x: matrix[query][x], reverse=True )]
    pd.DataFrame( tmp ).to_csv( query + '.csv', index=False, header=False )
    pprint( tmp )
```

```python
# if wondering the M most popular name in t type of entity, use this cell
# entity types: see above

M = int(input('how many? '))
t = input('which type of entity? ')

count_dict = Counter( df[df['type'] == t].iloc[:,2] )

count_df = sorted( count_dict, key=lambda x: count_dict[x], reverse=True)[:M]

count_df = pd.DataFrame( [ (i, count_dict[i]) for i in count_df ] )
count_df.to_csv('top' + str(N) + '-pop-org.csv', index=False, header=[t, 'counts'])
```

### more co-occurance
- nested query
- supported multiple values in each query
- this is based on all word-segment data, not only name entities

```python
# reload a file to a variable
with open('sentence_list.pickle', 'rb') as file:
    sent_list = pickle.load( file )
    file.close()
    
# print(len(sent_list))
# len(enr_list)
```

```python
first_query  = input('first query, seperate values with space: ').split()
second_query = input('second query, seperate values with space: ').split()

first_outcome  = []
second_outcome = {}
 
for i in sent_list:
    count = 0
    for keyword in first_query:
        if keyword in i:
            count += 1
    if count > 0:
        first_outcome.append(i)
        
print()
print('Simple Statistics: ')
print('In the articles that contains', first_query )
print('the number of articles contain second query are...')
for key in second_query:
    lst = [i for i in first_outcome if key in i]
    second_outcome[key] = lst
    print( key, len(lst) )
    
with open('output.txt', 'w', encoding='utf-8') as f:
    print(second_outcome, file=f)
    f.close()
    
print()
print('the nested output is sent to output.txt')


# readable output

second_outcome_readable = { key: [''.join(i) for i in second_outcome[key]] for key in second_outcome }
with open('output_whole_sentence.txt', 'w', encoding='utf-8') as f:
    pprint(second_outcome_readable, f)
    f.close()
    
print()
print('the readable nested output is sent to output.txt')
```

#### Appendix B: Data pre-preprocessing

```python
import pyodbc
print( [x for x in pyodbc.drivers() if x.startswith('Microsoft Access Driver')] )

conn_str = (
    r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
    r'DBQ=../留學.accdb;'
    )

# built connection
cnxn = pyodbc.connect(conn_str)
# cursor: the "mouse" in a connection
crsr = cnxn.cursor()

# if you do anything with the DB, remember to commit

for table_info in crsr.tables(tableType='TABLE'):
    print(table_info.table_name)
    
print(crsr.description)

# execute command with the crusor

crsr.execute('SELECT * from ts_page_content')
rows = crsr.fetchall()
    
crsr.close()
cnxn.close()

texts = np.array(rows)[:,-6]
```

```python
# load model
ws = WS("./data")
pos = POS("./data")
ner = NER("./data" ) #, disable_cuda=False)
```

```python
# build word segment output

sent_list = ws ( texts )
pos_list  = pos( sent_list )
```

```python
# save word-segment output

import pickle

# pickle a variable to a file
file = open('sentence_list.pickle', 'wb')
pickle.dump(sent_list, file)
file.close()
```

```python
# get ner output

ner_list = []
error_list = []

for i in range(len(sent_list)):
    
    try:
        ner_tmp = ner(sent_list[i: i+1], pos_list[i: i+1])
        ner_list += ner_tmp
        print(i, 'is done!')

    except AssertionError:
        error_list.append( i )
        print(i, 'has encoding issue, thus raise assertion error!')


file = open('NER_list.pickle', 'wb')
pickle.dump(ner_list, file)
file.close()

print(error_list, 'are excluded')
```
