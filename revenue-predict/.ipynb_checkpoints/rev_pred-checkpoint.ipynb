#### 讀取數據、整理數據

```r
data = read.csv('rev.csv') #讀取資料
```

```r
head(data) #查看表頭
```

```r
rev_vector = c(as.vector(as.numeric(sub(',', '', data[,2]))))
# inc_vector = c(as.vector(as.numeric(data[,3])))
#轉換資料型態成為ts()函數可以吃的向量
```

```r
rev_ts <- ts(rev(rev_vector), start=2007, frequency=4)
# inc_ts <- ts(rev(inc_vector), start=2007, frequency=4)
#轉換向量成為ts物件
```

#### 資料可視化

```r
autoplot(rev_ts) +
  ggtitle("CVS quarterly revenue") +
  xlab("Year") +
  ylab("million dollars")
```

```r
ggseasonplot(rev_ts, year.labels=TRUE, year.labels.left=TRUE) +
  ylab("million dollars") +
  ggtitle("Seasonal plot: CVS quarterly revenue")
#查看各季節成長
```

```r
ggsubseriesplot(rev_ts) +
  ylab("million dollars") +
  ggtitle("Seasonal subseries plot: CVS quarterly revenue")
#查看各季節不同年份下的均數
```

```r
gglagplot(rev_ts, main = 'Lagged PLot')
#递延项相关查看
```

```r
options(repr.plot.width=8, repr.plot.height=4) 
#jupyter notebook調整輸出圖片形狀的參數
ggAcf(rev_ts, main = 'ACF plot')
# t與 t-1 高度正相關
# 無明顯季節性
```

#### 切割訓練集與驗證集

```r
length(rev_ts)
```

```r
train = window(rev_ts, start=2007, end = c(2017, 4)) #測試集
test = window(rev_ts, start=2018) #近一年數據做為驗證集
```

```r
length(train)
length(test)
```

#### 簡易預測

```r
meanf(rev_ts, 4)
#average method
```

```r
naive(rev_ts, 4)
#naive method
```

```r
snaive(rev_ts, 4)
#seasonal naive method
```

```r
rwf(rev_ts, 4, drift = TRUE)
#drift method
```

```r
options(repr.plot.width=8, repr.plot.height=7)
autoplot(rev_ts) +
  autolayer(meanf(train, h=5),
    series="Mean", PI=FALSE) +
  autolayer(naive(train, h=5),
    series="Naïve", PI=FALSE) +
  autolayer(snaive(train, h=5),
    series="Seasonal naïve", PI=FALSE) +
  autolayer(rwf(train, h=4, drift = TRUE),
    series="Drift", PI=FALSE) +
  ggtitle("Forecasts for CVS quarterly revenue") +
  xlab("Year") + ylab("Million Dollars") +
  guides(colour=guide_legend(title="Forecast"))
#简易预测作图
```

#### 時間序列分解

```r
autoplot(rev_ts, series="Data") +
  autolayer(ma(rev_ts, 5), series="5-MA") +
  xlab("Year") + ylab("million dollars") +
  ggtitle("CVS quarterly revenue") +
  scale_colour_manual(values=c("Data"= "grey50","5-MA"= "red"),
                      breaks=c("Data", "5-MA"))
#移动平均法
```

```r
#傳統方法：
rev_ts %>% decompose(type="multiplicative") %>%
  autoplot() + xlab("Year") +
  ggtitle("Classical multiplicative decomposition of CVS revenue")
```

```r
#X11
options(repr.plot.width=3, repr.plot.height=6)
rev_ts %>% seas(x11="") -> fit
autoplot(fit) +
  ggtitle("X11") + xlab('Year')
```

```r
#SEATS
rev_ts %>% seas() %>%
autoplot() +
  ggtitle("SEATS") + xlab('Year')
```

```r
#STL
autoplot(stl(rev_ts, s.window = 13)) + 
    ggtitle('STL') + 
    xlab('Year')
```

```r
?stl #查詢函數細節
```

```r
options(repr.plot.width=4, repr.plot.height=3.5)
fit_stl_naive = stlf(train, method='naive', h=5) #STL預測 #method可為arima、naive、ets、rwdrift
autoplot(fit_stl_naive) + 
    autolayer(rev_ts, series = 'Data') +
    ylab('million dollars') + xlab('Year') + ggtitle('STL + Naive')
```

```r
fit_stl_drift = stlf(train, method='rwdrift', h=5) #STL預測 #method可為arima、naive、ets、rwdrift
autoplot(fit_stl_drift) + 
    autolayer(rev_ts, series = 'Data') +
    ylab('million dollars') + xlab('Year') + ggtitle('STL + Random Walk with Drift')
```

```r
options(repr.plot.width=4, repr.plot.height=3.5)
fit_stl_arima = stlf(train, method='arima', h=5) #STL預測 #method可為arima、naive、ets、rwdrift
autoplot(fit_stl_arima) + 
    autolayer(rev_ts, series = "Data") +
    ylab('million dollars') + xlab('Year') + ggtitle('STL + ARIMA(0,1,0) with drift')
```

```r
fit_stl_ets = stlf(train, method='ets', h=5) #STL預測 #method可為arima、naive、ets、rwdrift
autoplot(fit_stl_ets) + 
    autolayer(rev_ts, series = "Data") +
    ylab('million dollars') + xlab('Year') + ggtitle('STL + ETS(A,A,N)')
```

#### 指數平滑：處理帶有趨勢的資料

```r
options(repr.plot.width=8, repr.plot.height=4)
#簡單指數平滑
fit_ses = ses(train, h= 5)
autoplot(fit_ses) + 
  autolayer(rev_ts, series='Data') +
  autolayer(fitted(fit_ses), series="Fitted") +
  ylab("million dollars") + xlab("Year")
```

```r
options(repr.plot.width=8, repr.plot.height=4)
#假設趨勢固定，y = level + trend
fit_holt = holt(train, h = 5)
#若 phi < 1，假設趨勢遞減
fit_holt_damp = holt(train, damped = TRUE, phi = 0.9, h = 5)
autoplot(rev_ts) +
  autolayer(fit_holt, series="Holt's method", PI=FALSE) +
  autolayer(fit_holt_damp, series="Damped Holt's method", PI=FALSE) +
  autolayer(rev_ts, series = 'Data') +
  ggtitle("Forecasts from Holt's method") + xlab("Year") +
  ylab("million dollars") +
  guides(colour=guide_legend(title="Forecast"))
```

```r
#在holt中引入季節性，分成乘法以及加法
hw_fit1 <- hw(train, seasonal="additive", h=5)
hw_fit2 <- hw(train, seasonal="multiplicative", h=5)
autoplot(rev_ts) +
  autolayer(hw_fit1, series="HW additive forecasts", PI=FALSE) +
  autolayer(hw_fit2, series="HW multiplicative forecasts", PI=FALSE) +
  xlab("Year") +
  ylab("million dollars") +
  ggtitle("CVS quarterly revenue") +
  guides(colour=guide_legend(title="Forecast"))
```

```r
#ETS model
fit_ets = ets(train)
summary(fit_ets)
autoplot(fit_ets)
```

```r
fit_ets %>% forecast(h = 5) %>%
  autoplot() +
  autolayer(rev_ts, series='Data') + 
  ylab("million dollars") + xlab("Year")
```

```r
checkresiduals(fit_ets)
```

#### 其它方法

```r
fit_tbats = tbats(train)
#TBATS：允許資料的季節性隨時間變換
```

```r
fit_tbats %>% forecast(h = 5) %>%
  autoplot() +
  autolayer(rev_ts, series='Data') + 
  ylab("million dollars") + xlab("Year")
```

```r
checkresiduals(fit_tbats)
```

```r
#neural network
fit_nn = nnetar(train)
autoplot(forecast(fit_nn), h=5, series = 'Predict') +
    autolayer(rev_ts, series = 'Data') + ylab('million dollars') + xlab('year')
```

```r
fit_final = arima(rev_ts, order=c(1,1,1), include.mean = TRUE)
```
